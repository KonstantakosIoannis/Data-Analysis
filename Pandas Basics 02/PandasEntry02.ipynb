```python
alpha=beta[['Status','Location','height_ft','Manufacturer','Gforce_clean','Inversions_clean','coaster_name','year_introduced','latitude','longitude','Type_Main','speed_mph','opening_date_clean']]
```

```python
alpha=alpha.drop(['Gforce_clean'],axis=1)
```

```python
alpha.drop('latitude',axis=1,inplace=True)
```

```python
alpha=alpha.rename(columns={'coaster_name':'name','speed_mph':'speed'})
```

```python
alpha = alpha.loc[~alpha.duplicated(subset=['name','Location','opening_date_clean'])] .reset_index(drop=True)
```

```python
alpha[alpha['Location']!='Other'].groupby('Location')['speed'].mean().sort_values(ascending=False)
#first i drop the values with location=other
#2nd step group by location
#3rd step speed.mean 
#and finally i order  in descending
```

```python
alpha[alpha['Location']!='Other'].groupby('Location')['speed'].agg(['mean','count']).dropna().sort_values('count',ascending=False)
#here we add some more steps
#agg = when i want more than one function(eg. mean and count )
#i also dropped the Null values
```

```python
alpha[alpha['Location']!='Other'].groupby('Location')['speed'].agg(['mean','count']).query('count>=10').dropna().sort_values('mean',ascending=False)['mean']
```

```python
ax = alpha.query('Location != "Other"') \
    .groupby('Location')['speed'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='Average Coast Speed by Location')
ax.set_xlabel('Average Coaster Speed')
matplotlib.pyplot.show()
```
